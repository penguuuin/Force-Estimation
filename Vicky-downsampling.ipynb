```python
import tensorflow as tf
import pandas as pd
import numpy as np
import skimage
import matplotlib.pyplot as plt
import matplotlib
import pathlib
import PIL
import PIL.Image
from skimage.measure import block_reduce
from skimage import data, color
from skimage.transform import rescale
```

```python
color_dir = "./image_data/test-1"
data_dir = pathlib.Path(color_dir)
image_count = len(list(data_dir.glob('*.png')))

samples = list(data_dir.glob('*'))[:10]
```

```python
str(samples[0])
```

```python
I = np.array(PIL.Image.open(str(samples[0])))
img = PIL.Image.fromarray(I, 'RGB')
```

```python
img = np.asanyarray(img)
# img = color.rgb2gray(img)
image_rescaled = rescale(img, 0.5, anti_aliasing=False, multichannel = True)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

ax = axes.ravel()

ax[0].imshow(img)
ax[0].set_title("Original image")

ax[1].imshow(image_rescaled)
ax[1].set_title("Rescaled image (aliasing)")


plt.tight_layout()
plt.show()
```
